Fake News Detection
Built a model to predict the news is true or fake
Data collected from kaggle to train the model
Used Countvectorizer and MultinomialNB in training the model
Deployed the model using python library Streamlit

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_testing=pd.DataFrame()
```

```python
df_fake=pd.read_csv("/content/Fake.csv")

df_fake
```

```python
df_fake.shape
```

```python
df_fake=df_fake.drop(['title','subject','date'],axis="columns")
df_fake['detect']=0
df_fake
```

```python
df_fake.isnull().sum()
```

```python
df_true=pd.read_csv("/content/True.csv")
df_true
```

```python
df_true.shape
```

```python
df_true.isnull().sum()
```

```python
df_true=df_true.drop(['title','subject','date'],axis="columns")
df_true['detect']=1
df_true
```

```python
df_news=pd.concat([df_fake,df_true])
df_news
```

```python
df_news['text']=df_news['text'].str.lower()
df_news['text']=df_news['text'].str.replace('[^\w\s]', '')
df_news['text']=df_news['text'].str.replace('\d+', '')

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df_news.text,df_news.detect,test_size=0.25)
```

```python
y_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer(stop_words='english')
x_train_count=v.fit_transform(x_train)
```

```python
x_test_count=v.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train_count,y_train)
```

```python
model.score(x_test_count,y_test)
```

```python
data=["a gold bucket found in uk."]
data=v.transform(data)
model.predict(data)
```

```python
y_pred=model.predict(x_test_count)
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.pipeline import make_pipeline
model2=make_pipeline(CountVectorizer(),MultinomialNB())
model2.fit(x_train,y_train)
y_pred2=model2.predict(x_test)
```

```python
import joblib
joblib.dump(model2,'Fake News Detect')
```

```python
reloaded_model=joblib.load('Fake News Detect')
reloaded_model
```
